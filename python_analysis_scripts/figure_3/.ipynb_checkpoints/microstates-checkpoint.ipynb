```python
# Libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def Q(pH, pka1, pka2, W):
    return 1 + 10**((pka1-pH)) + 10**((pka2-pH)) + 10**((pka1+pka2-W-2*pH))

def f00(pH, pka1, pka2, W):
    return 1/Q(pH, pka1, pka2, W)

def f01(pH, pka1, pka2, W):
    return 10**((pka1-pH))/Q(pH, pka1, pka2, W)

def f10(pH, pka1, pka2, W):
    return 10**((pka2-pH))/Q(pH, pka1, pka2, W)

def f11(pH, pka1, pka2, W):
    return 10**((pka1+pka2-W-2*pH))/Q(pH, pka1, pka2, W)

# Fit all data simultaneously
def combined_func(x, pka1, pka2, W):
    pH = x
    return np.concatenate([f00(pH, pka1, pka2, W), 
                           f01(pH, pka1, pka2, W), 
                           f10(pH, pka1, pka2, W), 
                           f11(pH, pka1, pka2, W)])
```

```python
data_pathf="../data/CpHMD/biliverdin.water/simultaneous"
#data_pathf="../data/monomer.complex/"
xf = np.arange(1.0,10.0,0.5)
x01f = np.loadtxt(f'{data_pathf}/p01.dat')
x10f = np.loadtxt(f'{data_pathf}/p10.dat')
x00f = np.loadtxt(f'{data_pathf}/p00.dat')
x11f = np.loadtxt(f'{data_pathf}/p11.dat')

X_t = x01f + x10f + 2*x11f
np.set_printoptions(suppress=True, precision=5)
print(2-X_t)
```

```python
y_sim = np.concatenate([x00f, x01f, x10f, x11f])

initial_guess = [7, 7, 0] 
params, covariance = curve_fit(combined_func, xf, y_sim, p0=initial_guess)

pka1 = params[0]
pka2 = params[1]
W = params[2]

print(f"pka1=params[0] is {pka1:.4f}")
print(f"pka2=params[1] is {pka2:.4f}")
print(f"W=params[2] is {W:.4f}")
```

```python
plt.figure(figsize=(12, 8))

np.seterr(divide='ignore') # Ignore the warnings that come from taking the log of tiny numbers
xn=np.arange(1,10.01,0.01)

plt.scatter(xf,x00f, label=r'$p_{00}(pH)$', color='blue', s=500, alpha=0.7, edgecolor='blue', linewidth=1.5)
plt.plot(xn, f00(xn,pka1,pka2,W), color='LightBlue', lw=6)
         
#plt.scatter(xf,x10f+x01f, color='red', label=r'$p_{01}(pH)+p_{10}(pH)$', s=500, alpha=0.7, edgecolor='red', linewidth=1.5)
#plt.plot(xn, f10(xn,pka1,pka2,W)+f01(xn,pka1,pka2,W), color='Salmon', lw=6)

plt.scatter(xf,x10f, color='red', label=r'$p_{10}(pH)$', s=500, alpha=0.7, edgecolor='red', linewidth=1.5)
plt.plot(xn, f10(xn,pka1,pka2,W), color='Salmon', lw=6)
plt.scatter(xf,x01f, color='black', label=r'$p_{01}(pH)$', s=500, alpha=0.7, edgecolor='black', linewidth=1.5)
plt.plot(xn, f01(xn,pka1,pka2,W), color='grey', lw=6)

plt.scatter(xf,x11f, color='green', label=r'$p_{11}(pH)$', s=500, alpha=0.7, edgecolor='green', linewidth=1.5)
plt.plot(xn,f11(xn,pka1,pka2,W), color='Lightgreen', lw=6)


#plt.axvline(x=3.9, ymin=0, ymax=0.5, color='black', linestyle='--', lw=4)
#plt.axvline(x=4.63, ymin=0, ymax=0.145, color='black', linestyle='--', lw=4)
#plt.axvline(x=5.3, ymin=0, ymax=0.5, color='black', linestyle='--', lw=4)
#plt.axhline(y=0, color='black', linestyle='-', lw=2)

plt.xlim(2,8)
plt.ylim(0,1)
plt.xlabel('pH', fontsize='40')

#plt.title('Total titration curve and titration curves of the two sites', fontsize=20)

ax = plt.gca()  # Get the current axes
line_width = 3
ax.spines['top'].set_linewidth(line_width)
ax.spines['bottom'].set_linewidth(line_width)
ax.spines['left'].set_linewidth(line_width)
ax.spines['right'].set_linewidth(line_width)

x_ticks = [2, 3, 3.9, 4.63, 5.3, 6, 7, 8]
x_labels = [2, 3, 3.9, 4.6, 5.3, 6, 7, 8]
plt.xticks(x_ticks, x_labels)
y_ticks = [0, 0.25, 0.5, 0.75, 1]
y_labels = [0, 0.25, 0.5, 0.75, 1]
plt.yticks(y_ticks, y_labels)
plt.tick_params(labelsize=30) 

plt.legend(prop={'size': 20}, frameon=False, loc='center right')
plt.grid(True)
plt.show()
#plt.savefig(fr'./microstates.svg', format='svg', bbox_inches='tight')
```

```python
data_pathm="../data/CpHMD/monomer.complex"
xm = np.arange(3.5,10.5,0.5)
x01m = np.loadtxt(f'{data_pathm}/p01.dat')
x10m = np.loadtxt(f'{data_pathm}/p10.dat')
x00m = np.loadtxt(f'{data_pathm}/p00.dat')
x11m = np.loadtxt(f'{data_pathm}/p11.dat')

def Qm(pH, pka1, pka2, W, n1, n2, n3):
    return 1 + 10**(n1*(pka1-pH)) + 10**(n2*(pka2-pH)) + 10**(n3*(pka1+pka2-W-2*pH))

def f00m(pH, pka1, pka2, W, n1, n2, n3):
    return 1/Qm(pH, pka1, pka2, W, n1, n2, n3)

def f01m(pH, pka1, pka2, W, n1, n2, n3):
    return 10**(n1*(pka1-pH))/Qm(pH, pka1, pka2, W, n1, n2, n3)

def f10m(pH, pka1, pka2, W, n1, n2, n3):
    return 10**(n2*(pka2-pH))/Qm(pH, pka1, pka2, W, n1, n2, n3)

def f11m(pH, pka1, pka2, W, n1, n2, n3):
    return 10**(n3*(pka1+pka2-W-2*pH))/Qm(pH, pka1, pka2, W, n1, n2, n3)

# Fit all data simultaneously
def combined_func_m(x, pka1, pka2, W, n1, n2, n3):
    pH = xm
    return np.concatenate([f00m(pH, pka1, pka2, W, n1, n2, n3), 
                           f01m(pH, pka1, pka2, W, n1, n2, n3), 
                           f10m(pH, pka1, pka2, W, n1, n2, n3), 
                           f11m(pH, pka1, pka2, W, n1, n2, n3)])

y_m = np.concatenate([x00m, x01m, x10m, x11m])

initial_guess = [7, 7, 0, 1, 1, 1] 
params, covariance = curve_fit(combined_func_m, xm, y_m, p0=initial_guess)

pka1m = params[0]
pka2m = params[1]
Wm = params[2]
n1m = params[3]
n2m = params[4]
n3m = params[5]

print(f"pka1=params[0] is {pka1m:.4f}")
print(f"pka2=params[1] is {pka2m:.4f}")
print(f"W=params[2] is {Wm:.4f}")
print(f"n1=params[3] is {n1m:.4f}")
print(f"n2=params[4] is {n2m:.4f}")
print(f"n3=params[5] is {n3m:.4f}")
```

```python
X_m = x01m + x10m + 2*x11m
np.set_printoptions(suppress=True, precision=5)
print(2-X_m)
```

```python
plt.figure(figsize=(12, 8))

np.seterr(divide='ignore') # Ignore the warnings that come from taking the log of tiny numbers
xk=np.arange(1,10.51,0.01)

plt.scatter(xm,x00m, label=r'$p_{00}(pH)$', color='blue', s=500, alpha=0.7, edgecolor='blue', linewidth=1.5)
plt.plot(xk, f00m(xk,pka1m,pka2m,Wm,n1m,n2m,n3m), color='LightBlue', lw=6)
         
#plt.scatter(xm,x10m+x01m, color='red', label=r'$p_{01}(pH)+p_{10}(pH)$', s=500, alpha=0.7, edgecolor='red', linewidth=1.5)
#plt.plot(xk, f10m(xk,pka1m,pka2m,Wm,n1m,n2m,n3m)+f01m(xk,pka1m,pka2m,Wm,n1m,n2m,n3m), color='Salmon', lw=6)

plt.scatter(xm,x10m, color='red', label=r'$p_{10}(pH)$', s=500, alpha=0.7, edgecolor='red', linewidth=1.5)
plt.plot(xk, f10m(xk,pka1m,pka2m,Wm,n1m,n2m,n3m), color='Salmon', lw=6)
plt.scatter(xm,x01m, color='black', label=r'$p_{01}$', s=500, alpha=0.7, edgecolor='black', linewidth=1.5)
plt.plot(xk, f01m(xk,pka1m,pka2m,Wm,n1m,n2m,n3m), color='grey', lw=6)

plt.scatter(xm,x11m, color='green', label=r'$p_{11}(pH)$', s=500, alpha=0.7, edgecolor='green', linewidth=1.5)
plt.plot(xk,f11m(xk,pka1m,pka2m,Wm,n1m,n2m,n3m), color='Lightgreen', lw=6)

#plt.axvline(x=6.76, ymin=0, ymax=0.5, color='black', linestyle='--', lw=4)


plt.xlim(3.5,10)
plt.ylim(0,1)
plt.xlabel('pH', fontsize='40')

#plt.title('Total titration curve and titration curves of the two sites', fontsize=20)

ax = plt.gca()  # Get the current axes
line_width = 3
ax.spines['top'].set_linewidth(line_width)
ax.spines['bottom'].set_linewidth(line_width)
ax.spines['left'].set_linewidth(line_width)
ax.spines['right'].set_linewidth(line_width)

#x_ticks = [3, 4, 5, 6, 7, 8, 9, 10]
#x_labels = [3, 4, 5, 6, 7, 8, 9, 10]
#plt.xticks(x_ticks, x_labels)
#y_ticks = [0, 0.25, 0.5, 0.75, 1]
#y_labels = [0, 0.25, 0.5, 0.75, 1]
#plt.yticks(y_ticks, y_labels)
plt.tick_params(labelsize=30) 

#plt.legend(prop={'size': 25}, frameon=False, loc='center right')
plt.grid(True)
plt.show()
#plt.savefig(fr'./microstates.svg', format='svg', bbox_inches='tight')
```
