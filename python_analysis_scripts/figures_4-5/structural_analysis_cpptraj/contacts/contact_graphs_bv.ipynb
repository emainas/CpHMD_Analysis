```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
monomer = np.genfromtxt('./monomer/BV_5Ang.gnu', skip_header=7, skip_footer=2)
x1 = monomer[:, 0]  
y1 = monomer[:, 1]  
z1 = monomer[:, 2]

tet_chainA_sameunit = np.genfromtxt('./tetramer/chainA/BV339_chainA_5Ang.gnu', skip_header=7, skip_footer=2)
tet_chainA_interface = np.genfromtxt('./tetramer/chainA/BV339_chainC_5Ang.gnu', skip_header=7, skip_footer=2)
x_tet_chainA_sameunit = tet_chainA_sameunit[:, 0]  
y_tet_chainA_sameunit = tet_chainA_sameunit[:, 1]  
z_tet_chainA_sameunit = tet_chainA_sameunit[:, 2]
x_tet_chainA_interface = tet_chainA_interface[:, 0]  
y_tet_chainA_interface = tet_chainA_interface[:, 1]  
z_tet_chainA_interface = tet_chainA_interface[:, 2]

tet_chainB_sameunit = np.genfromtxt('./tetramer/chainB/BV340_chainB_5Ang.gnu', skip_header=7, skip_footer=2)
tet_chainB_interface = np.genfromtxt('./tetramer/chainB/BV340_chainD_5Ang.gnu', skip_header=7, skip_footer=2)
x_tet_chainB_sameunit = tet_chainB_sameunit[:, 0]  
y_tet_chainB_sameunit = tet_chainB_sameunit[:, 1]  
z_tet_chainB_sameunit = tet_chainB_sameunit[:, 2]
x_tet_chainB_interface = tet_chainB_interface[:, 0]  
y_tet_chainB_interface = tet_chainB_interface[:, 1]  
z_tet_chainB_interface = tet_chainB_interface[:, 2]

tet_chainC_sameunit = np.genfromtxt('./tetramer/chainC/BV679_chainC_5Ang.gnu', skip_header=7, skip_footer=2)
tet_chainC_interface = np.genfromtxt('./tetramer/chainC/BV679_chainA_5Ang.gnu', skip_header=7, skip_footer=2)
x_tet_chainC_sameunit = tet_chainC_sameunit[:, 0]  
y_tet_chainC_sameunit = tet_chainC_sameunit[:, 1]  
z_tet_chainC_sameunit = tet_chainC_sameunit[:, 2]
x_tet_chainC_interface = tet_chainC_interface[:, 0]  
y_tet_chainC_interface = tet_chainC_interface[:, 1]  
z_tet_chainC_interface = tet_chainC_interface[:, 2]

tet_chainD_sameunit = np.genfromtxt('./tetramer/chainD/BV680_chainD_5Ang.gnu', skip_header=7, skip_footer=2)
tet_chainD_interface = np.genfromtxt('./tetramer/chainD/BV680_chainB_5Ang.gnu', skip_header=7, skip_footer=2)
x_tet_chainD_sameunit = tet_chainD_sameunit[:, 0]  
y_tet_chainD_sameunit = tet_chainD_sameunit[:, 1]  
z_tet_chainD_sameunit = tet_chainD_sameunit[:, 2]
x_tet_chainD_interface = tet_chainD_interface[:, 0]  
y_tet_chainD_interface = tet_chainD_interface[:, 1]  
z_tet_chainD_interface = tet_chainD_interface[:, 2]
```

```python
x1u = np.unique(x1)
y1u = np.unique(y1)
Z1 = z1.reshape(len(y1u), len(x1u))

x2u = np.unique(x_tet_chainA_sameunit)
y2u = np.unique(y_tet_chainA_sameunit)
Z2 = z_tet_chainA_sameunit.reshape(len(y2u), len(x2u))
x3u = np.unique(x_tet_chainA_interface)
y3u = np.unique(y_tet_chainA_interface)
Z3 = z_tet_chainA_interface.reshape(len(y3u), len(x3u))

x4u = np.unique(x_tet_chainB_sameunit)
y4u = np.unique(y_tet_chainB_sameunit)
Z4 = z_tet_chainB_sameunit.reshape(len(y4u), len(x4u))
x5u = np.unique(x_tet_chainB_interface)
y5u = np.unique(y_tet_chainB_interface)
Z5 = z_tet_chainB_interface.reshape(len(y5u), len(x5u))

x6u = np.unique(x_tet_chainC_sameunit)
y6u = np.unique(y_tet_chainC_sameunit)
Z6 = z_tet_chainC_sameunit.reshape(len(y6u), len(x6u))
x7u = np.unique(x_tet_chainC_interface)
y7u = np.unique(y_tet_chainC_interface)
Z7 = z_tet_chainC_interface.reshape(len(y7u), len(x7u))

x8u = np.unique(x_tet_chainD_sameunit)
y8u = np.unique(y_tet_chainD_sameunit)
Z8 = z_tet_chainD_sameunit.reshape(len(y8u), len(x8u))
x9u = np.unique(x_tet_chainD_interface)
y9u = np.unique(y_tet_chainD_interface)
Z9 = z_tet_chainD_interface.reshape(len(y9u), len(x9u))
```

```python
y_mon = np.where(y1u == 170)[0][0]  
z_mon = Z1[y_mon, :]  

y339_chainA = np.where(y2u == 339)[0][0]  
z339_chainA = Z2[y339_chainA, :]  
y339_chainC = np.where(y3u == 339)[0][0]  
z339_chainC = Z3[y339_chainC, :]  

y340_chainA = np.where(y4u == 340)[0][0]  
z340_chainA = Z4[y340_chainA, :]  
y340_chainC = np.where(y5u == 340)[0][0]  
z340_chainC = Z5[y340_chainC, :]  

y679_chainA = np.where(y6u == 679)[0][0]  
z679_chainA = Z6[y679_chainA, :]  
y679_chainC = np.where(y7u == 679)[0][0]  
z679_chainC = Z7[y679_chainC, :]  

y680_chainA = np.where(y8u == 680)[0][0]  
z680_chainA = Z8[y680_chainA, :]  
y680_chainC = np.where(y9u == 680)[0][0]  
z680_chainC = Z9[y680_chainC, :]  
```

```python
fig, (ax1, ax2) = plt.subplots(2, 2, figsize=(15, 10))

#ax1[0].plot(x1u, z_mon, marker='o', lw=0.2, markersize=6, color='red', label='Monomer')
ax1[0].plot(x2u, z339_chainA/np.max(z339_chainA), marker='o', lw=0.2, markersize=6, color='Lightgreen', label='Tetramer BV339 with residues in Chain-A ')
ax1[0].plot(x3u, z339_chainC/np.max(z339_chainA), marker='o', lw=0.2, markersize=6, color='red', label='Tetramer BV339 with residues in Chain-C')
ax1[0].set_xlabel('residue number')
ax1[0].set_ylabel(r'Prob. relative to closest res. (PHE36)')
ax1[0].set_title(f'Heavy atom contacts of BV339 within 5 Angstrom')
ax1[0].grid(True)
#plt.xlim(0,150)
ax1[0].legend()
highlighted_x1 = [68, 60]
for x_val1 in highlighted_x1:
    if x_val1 in x2u:
        ax1[0].plot(x_val1, z339_chainA[x2u == x_val1] / np.max(z339_chainA), marker='o', lw=0.2, markersize=6, color='blue')
    if x_val1 in x3u:
        ax1[0].plot(x_val1, z339_chainC[x3u == x_val1] / np.max(z339_chainA), marker='o', lw=0.2, markersize=6, color='blue')


ax1[1].plot(x4u, z340_chainA/np.max(z340_chainA), marker='o', lw=0.2, markersize=6, color='Lightgreen', label='Tetramer BV340 with residues in Chain-B ')
ax1[1].plot(x5u, z340_chainC/np.max(z340_chainA), marker='o', lw=0.2, markersize=6, color='red', label='Tetramer BV340 with residues in Chain-D')
ax1[1].set_xlabel('residue number')
ax1[1].set_ylabel(r'Prob. relative to closest res. (PHE205)')
ax1[1].set_title(f'Heavy atom contacts of BV340 within 5 Angstrom')
ax1[1].grid(True)
ax1[1].legend()
#highlighted_x2 = [204, 207, 229, 237]
highlighted_x2 = [229, 237]
for x_val2 in highlighted_x2:
    if x_val2 in x4u:
        ax1[1].plot(x_val2, z340_chainA[x4u == x_val2] / np.max(z340_chainA), marker='o', lw=0.2, markersize=6, color='blue')
    if x_val2 in x5u:
        ax1[1].plot(x_val2, z340_chainC[x5u == x_val2] / np.max(z340_chainA), marker='o', lw=0.2, markersize=6, color='blue')


ax2[0].plot(x7u, z679_chainC/np.max(z679_chainA), marker='o', lw=0.2, markersize=6, color='red', label='Tetramer BV679 with residues in Chain-A')
ax2[0].plot(x6u, z679_chainA/np.max(z679_chainA), marker='o', lw=0.2, markersize=6, color='Lightgreen', label='Tetramer BV679 with residues in Chain-C ')
ax2[0].set_xlabel('residue number')
ax2[0].set_ylabel(r'Prob. relative to closest res. (PHE376)')
ax2[0].set_title(f'Heavy atom contacts of BV679 within 5 Angstrom')
ax2[0].grid(True)
ax2[0].legend()
#highlighted_x3 = [375, 378, 400, 408]
highlighted_x3 = [400, 408]
for x_val3 in highlighted_x3:
    if x_val3 in x6u:
        val = z679_chainA[x6u == x_val3] / np.max(z679_chainA)
        if val != 0:  
            ax2[0].plot(x_val3, val, marker='o', lw=0.2, markersize=6, color='blue')
    if x_val3 in x7u:
        val = z679_chainC[x7u == x_val3] / np.max(z679_chainA)
        if val != 0:  
            ax2[0].plot(x_val3, val, marker='o', lw=0.2, markersize=6, color='blue')


#ax2[1].plot(x1u, z_mon, marker='o', lw=0.2, markersize=6, color='red', label='Monomer')
ax2[1].plot(x9u, z680_chainC/np.max(z680_chainA), marker='o', lw=0.2, markersize=6, color='red', label='Tetramer BV680 with residues in Chain-B')
ax2[1].plot(x8u, z680_chainA/np.max(z680_chainA), marker='o', lw=0.2, markersize=6, color='Lightgreen', label='Tetramer BV680 with residues in Chain-D ')
ax2[1].set_xlabel('residue number')
ax2[1].set_ylabel(r'Prob. relative to closest res. (PHE545)')
ax2[1].set_title(f'Heavy atom contacts of BV680 within 5 Angstrom')
ax2[1].grid(True)
ax2[1].legend()
#highlighted_x4 = [544, 547, 569, 577]
highlighted_x4 = [569, 577]
for x_val4 in highlighted_x4:
    if x_val4 in x8u:
        val = z680_chainA[x8u == x_val4] / np.max(z680_chainA)
        if val != 0:  # Check if the value is not zero
            ax2[1].plot(x_val4, val, marker='o', lw=0.2, markersize=6, color='blue')
    if x_val4 in x9u:
        val = z680_chainC[x9u == x_val4] / np.max(z680_chainA)
        if val != 0:  # Check if the value is not zero
            ax2[1].plot(x_val4, val, marker='o', lw=0.2, markersize=6, color='blue')


max_index1 = np.argmax(z679_chainA)
x_value_at_max1 = x6u[max_index1]
print(f'The x value corresponding to the maximum value of z679_chainA is: {x_value_at_max1}')
max_index2 = np.argmax(z680_chainA)
x_value_at_max2 = x8u[max_index2]
print(f'The x value corresponding to the maximum value of z679_chainA is: {x_value_at_max2}')
max_index3 = np.argmax(z339_chainA)
x_value_at_max3 = x2u[max_index3]
print(f'The x value corresponding to the maximum value of z679_chainA is: {x_value_at_max3}')
max_index4 = np.argmax(z340_chainA)
x_value_at_max4 = x4u[max_index4]
print(f'The x value corresponding to the maximum value of z679_chainA is: {x_value_at_max4}')


plt.tight_layout()
plt.show()
```
