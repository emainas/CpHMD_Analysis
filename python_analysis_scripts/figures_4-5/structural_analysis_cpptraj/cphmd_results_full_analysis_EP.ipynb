```python
# We can measure the water molecules around the two chains of biliverdin to quantify how solvated they are. 

pAn_monomer = np.loadtxt('./hydration-calculations/watershells/monomer/chainA-waters.out', usecols=1, skiprows=0)
pDn_monomer = np.loadtxt('./hydration-calculations/watershells/monomer/chainD-waters.out', usecols=1, skiprows=0)

p339An_tetramer = np.loadtxt('./hydration-calculations/watershells/tetramer/chainA/chain_A-prnA-waters.out', usecols=1, skiprows=0)
p339Dn_tetramer = np.loadtxt('./hydration-calculations/watershells/tetramer/chainA/chain_A-prnD-waters.out', usecols=1, skiprows=0)

p340An_tetramer = np.loadtxt('./hydration-calculations/watershells/tetramer/chainB/chain_B-prnA-waters.out', usecols=1, skiprows=0)
p340Dn_tetramer = np.loadtxt('./hydration-calculations/watershells/tetramer/chainB/chain_B-prnD-waters.out', usecols=1, skiprows=0)

p679An_tetramer = np.loadtxt('./hydration-calculations/watershells/tetramer/chainC/chain_C-prnA-waters.out', usecols=1, skiprows=0)
p679Dn_tetramer = np.loadtxt('./hydration-calculations/watershells/tetramer/chainC/chain_C-prnD-waters.out', usecols=1, skiprows=0)

p680An_tetramer = np.loadtxt('./hydration-calculations/watershells/tetramer/chainD/chain_D-prnA-waters.out', usecols=1, skiprows=0)
p680Dn_tetramer = np.loadtxt('./hydration-calculations/watershells/tetramer/chainD/chain_D-prnD-waters.out', usecols=1, skiprows=0)

average_A_tetramer=(np.mean(p339An_tetramer)+np.mean(p340An_tetramer)+np.mean(p679An_tetramer)+np.mean(p680An_tetramer))/4
average_D_tetramer=(np.mean(p339Dn_tetramer)+np.mean(p340Dn_tetramer)+np.mean(p679Dn_tetramer)+np.mean(p680Dn_tetramer))/4
monomer_A=np.mean(pAn_monomer)
monomer_D=np.mean(pDn_monomer)

print(f"Monomer left side (prn-A): {monomer_A:.2f} and Tetramer average left-side (prn-A) is: {average_A_tetramer:.2f} ")
print(f"Monomer right side (prn-D): {monomer_D:.2f} and Tetramer average right-side (prn-D)is: {average_D_tetramer:.2f}")

data_pairs = [
    (pAn_monomer, pDn_monomer, 'Monomer A vs D'),
    (p339An_tetramer, p339Dn_tetramer, 'Tetramer A339 vs D339'),
    (p340An_tetramer, p340Dn_tetramer, 'Tetramer A340 vs D340'),
    (p679An_tetramer, p679Dn_tetramer, 'Tetramer A679 vs D679'),
    (p680An_tetramer, p680Dn_tetramer, 'Tetramer A680 vs D680')
]


fig, axes = plt.subplots(3, 2, figsize=(12, 12))
axes = axes.flatten()

for i, ax in enumerate(axes[:-1]):  # We use axes[:-1] to fill the first 5 subplots
    data_A, data_D, title = data_pairs[i]
    ax.hist(data_A, bins=30, color='blue', alpha=0.6, label=f'A (mean: {np.mean(data_A):.2f})')
    ax.hist(data_D, bins=30, color='red', alpha=0.6, label=f'D (mean: {np.mean(data_D):.2f})')
    ax.set_title(title)
    #sum+=np.mean(data_A)
    ax.legend()
    
fig.delaxes(axes[-1])
plt.tight_layout()
plt.show()
```

# 4) Intramolecular hydrogen bonding between prn chains in BV in the tetramer

```python
# I realized that we cannot read hydrogen bond statistics as they are because we have dummy atoms in a CpHMD so we end up
# OVERCOUNTING. We can look at the hydrogen bonding statistics as indication of a hydrogen bond but we cannot use the 
# average numbers computed in the cpptraj output files. The hydrogen bond analysis shows that in the tetramer, the two chains
# form hydrogen bonds way more ofter as compared to the monomer, a possible outcome of the more compact tetrameric structure.
# To quantify this and explain it I will use the prn-A -- prn-D distance as a proxy of hydrogen bonding and then I show how
# the underlying distribution can be split into protonated and deprotonated. 
```

# 5) Interesting interaction/hydrogen-bonding between BV339 in one Sandercyanin unit with an arginine from the neighboring unit. 

```python
data_arg = np.loadtxt('distance-arginine/bv339_prnA-arg347.out', skiprows=0)

x_arg = data_arg[:, 0]
y_arg = data_arg[:, 1]
x_arg=x_arg[:10000]
y_arg=y_arg[:10000]

y_zeros = [y_val for y_val, flag in zip(y_arg, prn_A54_10th[8]) if flag == 0]
y_ones = [y_val for y_val, flag in zip(y_arg, prn_A54_10th[8]) if flag == 1]

from scipy.optimize import curve_fit

def triple_gaussian(x, amp1, mean1, stddev1, amp2, mean2, stddev2, amp3, mean3, stddev3):
    gauss1 = amp1 * np.exp(-(x - mean1) ** 2 / (2 * stddev1 ** 2))
    gauss2 = amp2 * np.exp(-(x - mean2) ** 2 / (2 * stddev2 ** 2))
    gauss3 = amp3 * np.exp(-(x - mean3) ** 2 / (2 * stddev3 ** 2))
    return gauss1 + gauss2 + gauss3

y_zeros_hist, y_zeros_bin_edges = np.histogram(y_zeros, bins=90, density=True)
y_ones_hist, y_ones_bin_edges = np.histogram(y_ones, bins=90, density=True)
y_zeros_bin_centers = (y_zeros_bin_edges[:-1] + y_zeros_bin_edges[1:]) / 2
y_ones_bin_centers = (y_ones_bin_edges[:-1] + y_ones_bin_edges[1:]) / 2

initial_guess = [0.1, 4, 1, 0.5, 10, 1, 0.3, 9, 0.1]
popt_zeros, _ = curve_fit(triple_gaussian, y_zeros_bin_centers, y_zeros_hist, p0=initial_guess)
popt_ones, _ = curve_fit(triple_gaussian, y_ones_bin_centers, y_ones_hist, p0=initial_guess)


x_fit_zeros = np.linspace(min(y_zeros_bin_centers), max(y_zeros_bin_centers), 1000)
x_fit_ones = np.linspace(min(y_ones_bin_centers), max(y_ones_bin_centers), 1000)


plt.plot(x_fit_zeros, triple_gaussian(x_fit_zeros, *popt_zeros), color='brown', linestyle='-', linewidth=2, label='Deprotonated')
plt.plot(x_fit_ones, triple_gaussian(x_fit_ones, *popt_ones), color='gold', linestyle='-', linewidth=2, label='Protonated')

plt.title('Distance between BLA339-CGA and ARG347-CZ', fontsize=15)
plt.xlabel('r (Angstrom)', fontsize=18)
plt.ylabel('P(r)', fontsize=20)
plt.ylim(0,0.21)
plt.xlim(3.8,18)
plt.legend(fontsize=13)
plt.grid(False)

plt.tick_params(labelsize=15)
plt.tight_layout()
#plt.show()
plt.savefig('./arg_dist.svg', format='svg', bbox_inches='tight', dpi=300)


print('It seems that for a portion of the simulation, \
that the left part of biliverdin gets deprotonated - \
it interacts with ARG347 (a residue that is actually very far') 
```
