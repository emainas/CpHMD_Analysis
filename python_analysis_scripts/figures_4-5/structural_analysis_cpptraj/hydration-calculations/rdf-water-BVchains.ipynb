```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 5))

ax1.plot(r1a, g1a, color='green', label='O1A')
ax1.plot(r3a, g3a, color='red', label='O1D')
ax1.set_xlabel('r (Angstrom)')
ax1.set_xlim(0, 10)
ax1.legend(loc='best')
ax1.set_title('pH=4.0')

ax2.plot(r1n, g1n, color='green', label='O1A')
ax2.plot(r3n, g3n, color='red', label='O1D')
ax2.set_xlabel('r (Angstrom)')
ax2.set_xlim(0, 10)
ax2.set_ylabel('g(r)')
ax2.legend(loc='best')
ax2.set_title('pH=7.5')



plt.tight_layout()
plt.show()
```

```python
# Watershell analysis

watershell_path ='./watershells/pH4.0/'
pAa = np.loadtxt(watershell_path + 'chainA-waters.out', usecols=1, skiprows=0)
pDa = np.loadtxt(watershell_path + 'chainD-waters.out', usecols=1, skiprows=0)

watershell_path2 ='./watershells/pH7.5/'
pAn = np.loadtxt(watershell_path2 + 'chainA-waters.out', usecols=1, skiprows=0)
pDn = np.loadtxt(watershell_path2 + 'chainD-waters.out', usecols=1, skiprows=0)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(pAa, bins=50, color='green', alpha=0.6, label='O1A-O2A (pH 4.0)', density=True)
plt.hist(pDa, bins=50, color='red', alpha=0.6, label='O1D-O2D (pH 4.0)', density=True)
plt.title('pH 4.0')
plt.xlabel('Number of water molecules')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(pAn, bins=50, color='green', alpha=0.6, label='O1A-O2A (pH 7.5)', density=True)
plt.hist(pDn, bins=50, color='red', alpha=0.6, label='O1D-O2D (pH 7.5)', density=True)
plt.title('pH 7.5')
plt.xlabel('Number of water molecules')
plt.ylabel('Frequency')
plt.legend()


plt.tight_layout()
plt.show()


print('pH 4.0 Means:', np.mean(pAa), np.mean(pDa))
print('pH 7.5 Means:', np.mean(pAn), np.mean(pDn))
```
