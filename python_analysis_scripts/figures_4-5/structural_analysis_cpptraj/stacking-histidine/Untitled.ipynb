```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt('com-angle1.dat',skiprows=0)
r=data[:,1]
m=data[:,2]

```

```python
data_path=""
x = np.arange(3.5,10.5,0.5)
hip = np.loadtxt(f'{data_path}pHIP_res15.dat')
hid = np.loadtxt(f'{data_path}pHID_res15.dat')
hie = np.loadtxt(f'{data_path}pHIE_res15.dat')
```

```python
plt.figure(figsize=(6, 5))
plt.hist2d(r, m, bins=100, cmap='viridis', density=True)

plt.colorbar(label='Counts')
plt.xlabel('r (Angstrom)')
plt.ylabel('θ (degrees)')
plt.title('2D Histogram')

plt.tight_layout()
plt.show()
```

```python
hist, xedges, yedges = np.histogram2d(r, m, bins=100, density=True)

x_centers = (xedges[:-1] + xedges[1:]) / 2
y_centers = (yedges[:-1] + yedges[1:]) / 2
X, Y = np.meshgrid(x_centers, y_centers)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

Z = hist.T  # Transpose the histogram for correct orientation
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_xlabel('r (Angstrom)')
ax.set_ylabel('θ (degrees)')
ax.set_zlabel('Normalized Density')
ax.set_title('Surface Plot of Normalized Density')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

hist, xedges, yedges = np.histogram2d(r, m, bins=100, density=True)

x_centers = (xedges[:-1] + xedges[1:]) / 2
y_centers = (yedges[:-1] + yedges[1:]) / 2

X, Y = np.meshgrid(x_centers, y_centers)
Z = hist.T

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(X, Y, Z, edgecolor='none', antialiased=True)

ax.set_xlabel('r (Angstrom)')
ax.set_ylabel('θ (degrees)')
ax.set_zlabel('Normalized Density')
ax.set_title('Smooth 3D Surface Plot of Normalized Density')

ax.view_init(elev=30, azim=210)
plt.show()
```
