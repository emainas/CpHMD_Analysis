```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Histidines that are close to Biliverdins in the 4 tetramer chains

data_path=""
x = np.arange(3.5,10.5,0.5)
cHIPa = np.loadtxt(f'{data_path}pHIP_res15.dat')
cHIDa = np.loadtxt(f'{data_path}pHID_res15.dat')
cHIEa = np.loadtxt(f'{data_path}pHIE_res15.dat')

fHIPa = np.loadtxt(f'{data_path}pHIP_res19.dat')
fHIDa = np.loadtxt(f'{data_path}pHID_res19.dat')
fHIEa = np.loadtxt(f'{data_path}pHIE_res19.dat')

cHIPb = np.loadtxt(f'{data_path}pHIP_res42.dat')
cHIDb = np.loadtxt(f'{data_path}pHID_res42.dat')
cHIEb = np.loadtxt(f'{data_path}pHIE_res42.dat')

fHIPb = np.loadtxt(f'{data_path}pHIP_res46.dat')
fHIDb = np.loadtxt(f'{data_path}pHID_res46.dat')
fHIEb = np.loadtxt(f'{data_path}pHIE_res46.dat')

cHIPc = np.loadtxt(f'{data_path}pHIP_res71.dat')
cHIDc = np.loadtxt(f'{data_path}pHID_res71.dat')
cHIEc = np.loadtxt(f'{data_path}pHIE_res71.dat')

fHIPc = np.loadtxt(f'{data_path}pHIP_res75.dat')
fHIDc = np.loadtxt(f'{data_path}pHID_res75.dat')
fHIEc = np.loadtxt(f'{data_path}pHIE_res75.dat')

cHIPd = np.loadtxt(f'{data_path}pHIP_res98.dat')
cHIDd = np.loadtxt(f'{data_path}pHID_res98.dat')
cHIEd = np.loadtxt(f'{data_path}pHIE_res98.dat')

fHIPd = np.loadtxt(f'{data_path}pHIP_res102.dat')
fHIDd = np.loadtxt(f'{data_path}pHID_res102.dat')
fHIEd = np.loadtxt(f'{data_path}pHIE_res102.dat')
```

```python
fig, axs = plt.subplots(4, 2, figsize=(12, 12), sharey=True)  # 4 rows, 2 columns for 8 subplots
axs = axs.flatten()

axs[0].plot(x, cHIPa, label='HIP')
axs[0].plot(x, cHIDa, label='HID')
axs[0].plot(x, cHIEa, label='HIE')
axs[0].set_title('chain A - HIS 89')
axs[0].legend()

axs[1].plot(x, fHIPa, label='HIP')
axs[1].plot(x, fHIDa, label='HID')
axs[1].plot(x, fHIEa, label='HIE')
axs[1].set_title('chain A - HIS 120')
axs[1].legend()

axs[2].plot(x, cHIPb, label='HIP')
axs[2].plot(x, cHIDb, label='HID')
axs[2].plot(x, cHIEb, label='HIE')
axs[2].set_title('chain B - HIS 258')
axs[2].legend()

axs[3].plot(x, fHIPb, label='HIP')
axs[3].plot(x, fHIDb, label='HID')
axs[3].plot(x, fHIEb, label='HIE')
axs[3].set_title('chain B - HIS 289')
axs[3].legend()

axs[4].plot(x, cHIPc, label='HIP')
axs[4].plot(x, cHIDc, label='HID')
axs[4].plot(x, cHIEc, label='HIE')
axs[4].set_title('chain C - HIS 429')
axs[4].legend()

axs[5].plot(x, fHIPc, label='HIP')
axs[5].plot(x, fHIDc, label='HID')
axs[5].plot(x, fHIEc, label='HIE')
axs[5].set_title('chain C - HIS 460')
axs[5].legend()

axs[6].plot(x, cHIPd, label='HIP')
axs[6].plot(x, cHIDd, label='HID')
axs[6].plot(x, cHIEd, label='HIE')
axs[6].set_title('chain D - HIS 598')
axs[6].legend()

axs[7].plot(x, fHIPd, label='HIP')
axs[7].plot(x, fHIDd, label='HID')
axs[7].plot(x, fHIEd, label='HIE')
axs[7].set_title('chain D - HIS 629')
axs[7].legend()

plt.tight_layout()
#plt.show()
plt.savefig('histidines_in_tetramer.png')
```
